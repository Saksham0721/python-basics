```python
#Q.3 Compare and contrast mutable and immutable objects in Python with example.

#ANS.

#Mutable Objects :
#----------------

#These are the objects which when created can be manipulated at a later stage as per the requirement in our code. Any changes made to these are reflected in the orignal object itself. 

#some examples of mutable objects:

#(1)lists ==> A list is a very commonly used data structure in Python which can store any type of data in itself and is represented using square brackets []

#example:
list_of_fruits=['apple','orange','banana','mango']

#lists are mutable in nature i.e. the above list of fruits can be altered (adding a new fruit at the list end, changing an exisitng one or adding a new fruit after a specific index in the list)

list_of_fruits.append('papaya') #add new fruit to the end of list

list_of_fruits.insert(2,'grapes') # add new fruit after index 2 in the list

list_of_fruits[0] = 'guava'

#Output:

#['guava', 'orange', 'grapes', 'banana', 'mango', 'papaya']


#(2)Dictionary=> Dictionary in Python are key-value pairs which are represented by using curly braces{}.

#These are also mutable data structure in Python

my_phonebook = {'Dad':7865, 'Mum':4576, 'Friend':5668}

#Elements of a dicitionary can be accessed using corresponding keys to the values.

print(my_phonebook['Dad'])

#Output:

#7865

#we can also change the phone number corresponding to key 'Dad' and the changes made will be reflected in 'my_phonebook' [mutability]

my_phonebook['Dad']= '9999'

print([my_phonebook])

#Output:
#{'Dad': '9999', 'Mum': 4576, 'Friend': 5668}


#Immutable Objects:
#------------------

#These are the objects that are declared once can't be changed later. Any try to make a change in such objects will create a copy of original object in the memory with applied changes.

#some examples of immutable objects in Python:

#(1)Strings=> strings are a group of characters (alphabets, numbers, special character like @,$,# etc)  which are usually written in single quotes ('') or double quotes ("").The individual characters of a string can be accessed using indices starting from 0 to length of string - 1.

#These are immutable objects.

name = 'Rohan'

#name[0]='S' #this will simply raise an error that string literals do not support item assginment.

#However we can change this string as follows:

name.replace('R','S') # This will replace the alphabet 'R' in name with 'S' and output will be: 'Sohan'

#However the output generated is just a temporary copy of original string 'Rohan' with the applied changes.
#To verify this, we can print the variable 'name'

print(name)

#  Output:
# 'Rohan' i.e original string is retained in variable 'name' showcasing the immutability property of strings.


#(2) Tuples=> The tuples in Python are quite similar to lists but the major difference is that tuples are immutable in nature. These can also store any type of data and are represented by small brackets (). Tuples are very useful in the cases where manipulation of data,once it is stored, should be restricted.


#example=>
random_tuple=('saksham', 22.4, 1204, 3+5j, True,'Virat')

#random_tuple[1]='after' # error will be raised as tuples also do not support item assignment
```

```python
#Q.5 Explain the concept of type casting in Python with examples.

#ANS.
# Typecasting in Python is the process of converting one data type into another. 
# This is useful when you need to perform operations that require specific data types.
# 
# Typecasting in Python can be either implicit or explicit.
# 
# Implicit Typecasting:
# ---------------------
# Implicit typecasting is when Python automatically converts one data type to another without explicit instructions from the programmer. 
# This usually happens in operations where different data types are involved, and Python handles the conversion to avoid errors.
# 
# Example:
num_int = 10
num_float = 3.5

# Adding an integer and a float
result = num_int + num_float
print(result)  # Output: 13.5
print(type(result))  # Output: <class 'float'>

# In this example, num_int (an integer) is implicitly converted to a float before the addition.
# 
# Explicit Typecasting:
# ----------------------
# Explicit typecasting is when you manually convert one data type to another using built-in functions like int(), float(), str(), etc.
# 
# Example:
num_str = "123"
num_int = int(num_str)  # Explicitly converting string to integer
print(num_int)  # Output: 123
print(type(num_int))  # Output: <class 'int'>

# In this example, the string num_str is explicitly converted to an integer using the int() function.
```

```python
#Q.6 How do conditional statements work in Python? Illustrate with examples.

#ANS.
#In,Python the conditional statements are meant for decision-making. In, other words when execution of one or more statement relies on a condition then conditional statements are used.

#The main types of conditionals statements are:

#(1) if statement:
#-----------------
#Example: a python program to find even numbers from 1 to n.

n=int(input("Enter limit:"))# asking user to input limit upto which even numbers are to be printed

print(f" even numbers from 1 to {n} are:")
for i in range(1,n+1): # loop to iterate over sequence of numbers from 1 to n
    if(i%2==0):        #While iterating, this condition checks for every number 'i' in range 1 to n if it leaves                          #remainder zero when divided by 2.In this way, only those numbers which are even will get                          #printed.
        print(i,end=" ")

#(2) if-else statement:
#----------------------

#Example: Python program to check if two names are same or not.

name1=input("Enter 1st name:")
name2=input("Enter 2nd name:")

if(name1.lower()==name2.lower()):# name1 and name2 are converted to lowercase and then compared if they are same.
    print(f"Yes! the names {name1} and {name2} are same.")

else:# if the above 'if statement' does not get True then else statement will be executed.
    print(f"No! the names {name1} and {name2} are not same.")
       

#Output:

#Enter 1st name: Saksham
#Enter 2nd name: SAkSHam
#Yes! the names Saksham and SAkSHam are same.

#(3) if-elif-else statement (example for nested if-else as well):
#---------------------------------------------------------------

#Example: a python program to assign grade in maths based on the marks obtained by a student.

marks=int(input("enter your marks in maths out of 100:"))

if(marks<0 or marks>100):
    print("Kya kar raha hai bhai tu!")
    
else:

    if(marks>90 and marks<=100):#nested if-else
        if(marks==100):
            print("Marvellous :) ! you got an 'A+' in maths")
        else: 
            print("Congrats for achieving grade 'A' in maths")
        
    elif(marks>70 and marks<=90):
        print("Very Good! Grade 'B'")
        
    elif(marks>50 and marks<=70):
        print("Average! Grade 'C'")    
        
    elif(marks>=33 and marks<=50):
        print("Below Average!! Grade 'D' ")
        
    else:
        print("Extremely Poor!! Grade 'E' {FAIL}")


```

```python
#Q.7 Describe the different types of loops in Python and their use cases with examples.

#ANS.

#Loops in Python are extremely useful as they reduce the length of code, make our program more readable and organized.

#For loop in Python:
#-------------------

#It is used for iterating over a sequence of elements.

#Example: Python program to check how many times a number occurred in a list of numbers from 1 to 9

num_list=[1,2,5,4,7,3,6,9,9,9,9,9,7,2,2,8,8,8,8,3,2,2,2,2,1,1,6,5,4,6,5,4,3,3,5,5,7]

i = int(input("enter number whose frequency in the list is to be determined:"))

count = 0

for number in num_list: # this loop will go through each number of list 'num_list'
    if(number==i):# if the current number is equal to the number whose frequency is being checked then increase                     #variable count by 1
        count+=1
print(f"number {i} occurred {count} times in the above list")


#While loop in Python:
#---------------------
#While loop is used to execute one or more statements only until a certain condition is being met and once the condition becomes false, the execution of the statement is stopped.

#Example: python program to print all numbers divisible by 7  from 1 to num.

num=int(input("enter limit:"))
if(num>=1):
    i=1
    print(f"numbers divisible by 7 from 1 to {num} are:")
    while(i<=num):#allows if-statement to execute only until i is less than or equal to the limit entered
        if(i%7==0):
            print(i)
        i+=1   
        
else:
    print("Enter valid number!!")
    
    
#for-else loop:
#--------------

#The for-else loop in Python is a unique construct that allows you to execute a block of code when the loop completes normally (i.e., without encountering a break statement)


#Example:

numbers = [1, 2, 3, 4, 5]

for number in numbers:
    if number == 6:
        print("Found the number 6!")
        break
else:
    print("Number 6 not found in the list.")
    
#In this example, the else block runs because the loop goes through all the elements in the list without finding #the number 6.

#while-else loop:
#----------------

#Example:

count = 0

limit=int(input("enter limit:"))

while count < limit:
    if count ==5:
        print("Count reached 5, breaking out of the loop.")
        break
    count += 1
else:
    print("Loop completed normally, count never reached 5.")
    

#Output:

#enter limit: 4
#Loop completed normally, count never reached 5.
```
